```python
#### final interested columns
final_df = final_df[['sentID', 'sent', 'lexids', 'countWords']]
```

```python
final_df.head(), final_df.shape
```

```python
def get_noun_replacement(token):
    return nouns_dictionary[token].split()[1]
            
def process_sentence(sentences, lexids):
    
    final_sentences = []
    
    for sent, lexid in zip(sentences, lexids):
        temp = []
        tokens = sent.split(',')
        lexids = lexid.split(',')

        for i, token in enumerate(tokens):
            
            if (token == 'i' and lexids[i] != 'preverb'):
                temp.append(token)
                
            elif token in ['iic', 'i']:
                pass
            elif token in nouns_dictionary:
                temp.append(get_noun_replacement(token))
            elif token in verb_list:
                temp.append('head')
        
        final_sentences.append(temp)
    
    return final_sentences
```

```python
ans = process_sentence(final_df['sent'],  final_df['lexids'])
```
