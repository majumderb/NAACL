```python
import re
```

```python
fp = open('Hindi.txt' , 'r')
```

```python
data = fp.readlines()
```

```python
final_list = []

sentence_count = 0
cur_sentence = None
chunk_count = 0
exception_count =0 
for line in data:
    
    try:
        line = line.strip()

        ### check for sentence start and id 
        sent_start_pat= r"_<Sentence[\s_]+id=\'(\d+)\'>_"
        sent_start_obj = re.match(sent_start_pat, line)

        if sent_start_obj:
            cur_sentence = sent_start_obj.group(1)

            final_list.append({"chunks":[], "sentence_id":cur_sentence})
            print("sentence:"+cur_sentence+"started")

        ### pattern for extracting info about each chunk with drel


        chunk_drel_pat = r"[1-9]+\s*\(\([\s_]*([A-Z]+)[\s_]*<fs[\s_]*name=\'([A-Z1-9]+)\'[\s_]*drel=\'([a-z0-9-]+:[A-Z0-9]+)\'"
        chunk_drel_obj = re.match(chunk_drel_pat, line)

        if chunk_drel_obj:

    #         print("chunk")
    #         print(line)

    #         print("POS", chunk_drel_obj.group(1)) 
    #         print("name", chunk_drel_obj.group(2))
    #         print("drel", chunk_drel_obj.group(3))
    #         print("*"*10)        
            final_list[sentence_count]["chunks"].append({
                "pos":chunk_drel_obj.group(1),
                "name":chunk_drel_obj.group(2),
                "drel":chunk_drel_obj.group(3),
                "tokens":[]
            })
            chunk_count += 1



        ### pattern for extracting info about each chunk without drel
        chunk_voice_pat = r"[1-9]+\s*\(\([\s_]*([A-Z]+)[\s_]*<fs[\s_]*name=\'([A-Z1-9]+)\'[\s_]*(voicetype|comment)"
        chunk_voice_obj = re.match(chunk_voice_pat, line)
        if chunk_voice_obj:
    #         print("chunk")
    #         print("POS", chunk_voice_obj.group(1)) 
    #         print("name", chunk_voice_obj.group(2))
    #         print("*"*10)
            final_list[sentence_count]["chunks"].append({
                "pos":chunk_voice_obj.group(1),
                "name":chunk_voice_obj.group(2),
                "drel":"",
                "tokens":[]
            })
            chunk_count += 1



        ### pattern for extracting token info
        token_pat = r"[0-9]+\.[0-9]+\s+([a-zA-Z\.]+)[\s_]*([A-Z_]+)[\s_]*<fs[\s_]+af=\'([\.A-Za-z0-9,_]+)\'[\s_]*name=\'([A-Za-z_\.]+)\'>"
        token_obj = re.match(token_pat, line)
        if token_obj:
    #         print("token")
    #         print("POS", token_obj.group(2)) 
    #         print("af", token_obj.group(3))
    #         print("name", token_obj.group(4))
    #         print("*"*10)
#             print("chunk count", chunk_count)
            final_list[sentence_count]["chunks"][chunk_count-1]["tokens"].append({
                "pos":token_obj.group(2),
                "af":token_obj.group(3),
                "name":token_obj.group(4)
            })

        ### check for sentence start and id 
        sent_end_pat= r"_</Sentence>_"
        sent_end_obj = re.match(sent_end_pat, line)
        if sent_end_obj : 
            print("sentence end")
            sentence_count += 1
            chunk_count = 0 
    except:
        exception_count +=1 
        pass
    
```

```python
import json

with open('hindi_parsed_json.txt', 'w') as outfile:
    json.dump({'hindi':final_list}, outfile)
```
